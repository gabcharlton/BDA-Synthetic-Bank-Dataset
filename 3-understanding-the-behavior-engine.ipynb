# 🏦 BDA Synthetic Banking Simulator

# Understanding the Behaviour Engine

One of the goals of this project is **realism**.

Rather than generating completely random banking transactions, **the simulator attempts to mimic how real customers behave over time**.

In this notebook, we'll explore the ideas behind the Behaviour Engine.

---

## What is a Behaviour Engine?

The Behaviour Engine is responsible for answering questions like:

* Does this customer visit a branch today?
* Will they use an ATM?
* Are they likely to use online banking?
* What type of transaction will they perform?

Instead of making these decisions randomly, the simulator bases them on customer characteristics and external events.

---

## Customer Profiles

Every synthetic customer belongs to an underlying behavioural profile.

Examples include:

* Digital Native
* Hybrid Professional
* Traditional Branch User
* Senior Customer
* Small Business Owner
* High Net Worth Client
* Newcomer to Canada

These profiles influence the probability of different banking behaviours.

**Probabilities, Not Rules**

**Real people are unpredictable.**

Rather than saying:

_Seniors always visit branches._

The simulator says:

_Seniors are more likely to visit branches._

This probabilistic approach produces more realistic datasets.

---

## Digital Adoption

Every customer receives a Digital Adoption Score between 0 and 100.

Higher scores indicate greater comfort using digital banking services.

This score changes over time as customers gain experience with online banking.

---

## Pandemic Behaviour

The COVID-19 pandemic dramatically changed banking habits.

The simulator models several behavioural shifts:

* fewer branch visits
* increased online banking
* increased password resets
* increased phone banking
* gradual return to branches after restrictions eased

These changes occur gradually rather than instantly.

---

## Weather Effects

Weather also influences customer behaviour.

Examples include:

* snow reducing branch visits
* heavy rain encouraging digital banking
* windstorms affecting travel

These effects help create realistic fluctuations in daily activity.

---

## Sports and Entertainment

Because BDA is located in downtown Vancouver, nearby events influence banking behaviour.

Examples:

* Vancouver Canucks home games
* concerts
* festivals

Possible effects include:

* increased ATM usage
* larger cash withdrawals
* busier evenings
* Behaviour Over Time

Unlike static datasets, **customer behaviour evolves**.

_Digital confidence gradually increases._

_Branch dependency gradually decreases._

Customers also experience life events that may change their banking habits.

---

In this section we'll create simple charts showing:

These visualizations help verify that the simulator is behaving as expected.

---

## Summary

The Behaviour Engine combines:

* customer characteristics
* probabilities
* external events
* gradual behavioural change

to produce realistic banking activity over multiple years.
