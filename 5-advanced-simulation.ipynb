## Congratulations! 🥳

---

## Why Extend the Simulator?

Banks constantly evolve.

New products, technologies, regulations, and customer expectations change banking behaviour over time.

By extending the simulator, you can investigate new research questions.

---

## Adding New Customer Profiles

Consider creating additional behavioural groups such as:

* university students
* retirees
* frequent travellers
* newcomers to Canada
* gig economy workers

Each group may have different banking needs and digital preferences.

---

## Adding Life Events

Customers experience events that influence their financial behaviour.

Examples include:

* changing jobs
* moving to a new city
* marriage
* retirement
* starting a business
* purchasing a home

These events can trigger changes in account usage, transaction frequency, and channel preferences.

---

## Creating New Products

Modern banks offer a wide range of services.

Ideas include:

* investment accounts
* foreign currency accounts
* cryptocurrency services (simulated)
* youth accounts
* premium banking packages

Adding products creates new opportunities for analysis.

---

## Simulating Operational Events

Banks also experience operational disruptions.

Potential scenarios include:

* branch renovations
* ATM outages
* system maintenance
* internet outages
* severe weather
* public transit strikes

Researchers can study how these events influence customer behaviour.

---

## Scaling the Simulator

The modular design makes it possible to generate datasets ranging from a few thousand customers to hundreds of thousands.

When scaling up, consider:

* memory usage
* processing time
* storage requirements

Profiling and optimization become increasingly important.

---

## Future Research Ideas

This simulator can support many additional projects, including:

* branch staffing optimization
* fraud detection
* customer churn prediction
* financial inclusion research
* digital accessibility studies
* anomaly detection
* customer lifetime value modelling
* queue simulation
* synthetic data validation
