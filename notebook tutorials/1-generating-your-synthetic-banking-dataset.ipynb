# 🏦 BDA Synthetic Banking Simulator

# Generating Your First Banking Dataset

---

## Welcome!

In this notebook, you'll learn how to generate your own synthetic banking dataset using the BDA Banking Simulator.

By the end of this tutorial, you'll have created realistic banking data including:

* Customers
* Bank accounts
* Branches
* Calendar events
* Banking transactions

No banking data is real—everything is generated synthetically for research and educational purposes.

---

## Step 1: Import the Required Libraries

Every Python program begins by **importing** the tools it needs.

```python
import pandas as pd

from src.customers import generate_customers
from src.branches import generate_branches
from src.accounts import generate_accounts
from src.events import generate_calendar
from src.transactions import generate_transactions

```

_What just happened?_

We imported five generators.

| Generator	| What it creates |
| --- | --- |
| generate_customers()	| Customer information |
| generate_branches()	| Bank branch information |
| generate_accounts()	| Customer bank accounts |
| generate_calendar()	| Daily events like weather and holidays |
| generate_transactions()	| Banking activity |

---

## Step 2: Generate Customers

Let's create our customers.

```python
customers = generate_customers()

customers.head()

```

Expected output

```python
customer_id	first_name	age
100001	Emma	29
100002	James	63
...

```

_What happened?_

Each row represents one customer.

Every customer receives:

* a unique ID
* a synthetic name
* an age
* address information
* banking information

## Step 3: Save the Customer Dataset

Saving data allows us to reuse it later.

```python
customers.to_csv(
    "data/customers.csv",
    index=False
)
```

**Congratulations!**

You have created your first dataset. 😁

---

## Step 4: Generate Branches

Now let's generate the bank branches.

```python
branches = generate_branches()

branches
Save it
branches.to_csv(
    "data/branches.csv",
    index=False
)

```

---

## Step 5: Generate Accounts

Customers can own multiple bank accounts.

```python
accounts = generate_accounts(
    customers,
    branches
)

accounts.head()
```

_What happened?_

Each customer is assigned between one and several accounts.

Possible account types include

* Chequing
* Savings
* Credit Card
* Line of Credit
* Business Account

Save the accounts

```python
accounts.to_csv(
    "data/accounts.csv",
    index=False
)
```

---

## Step 6: Generate Calendar Events

The simulator includes external events that influence banking behaviour.

Generate the calendar.

```python
calendar = generate_calendar()

calendar.head()
```
Example

```python
Date	Weather	Holiday
2019-01-01	Snow	True
```

Save it

```python
calendar.to_csv(
    "data/calendar.csv",
    index=False
)
```

---

## Step 7: Generate Transactions

Now we can simulate banking activity.

```python
transactions = generate_transactions(
    customers,
    accounts,
    calendar
)

transactions.head()
```

This step may take several minutes depending on the number of customers.

_What is happening?_

The simulator is deciding

* Which customers visit the bank
* Which customers use the ATM
* Who uses online banking
* What transaction they perform
* Whether weather or holidays influence behaviour

---

## Step 8: Save the Transactions

```python
transactions.to_csv(
    "data/transactions.csv",
    index=False
)
```

**Excellent!** 🥳

You have generated a complete synthetic banking transaction dataset.

---

## Step 9: Check the Dataset Sizes

Let's see how much data was created.

```python
print("Customers:", len(customers))
print("Accounts:", len(accounts))
print("Calendar Days:", len(calendar))

```

## Step 10: Explore the Data

Pandas makes it easy to summarize information.

Average customer age

```python
customers["age"].describe()

```

Transaction channels

```python
transactions["channel"].value_counts()
```

Transaction types

---

## Step 11: Merge the Data

Because every dataset contains matching IDs, we can combine them.

Example
```python
customer_transactions = transactions.merge(
    customers,
    on="customer_id"
)

customer_transactions.head()
```
Now every transaction also contains customer information.

---

## Step 12: Save Everything Together

You may want all generated datasets in one folder.

```python
customers.to_csv("data/customers.csv", index=False)

branches.to_csv("data/branches.csv", index=False)

accounts.to_csv("data/accounts.csv", index=False)

calendar.to_csv("data/calendar.csv", index=False)

transactions.to_csv("data/transactions.csv", index=False)
```
