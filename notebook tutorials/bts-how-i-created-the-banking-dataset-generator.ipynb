# 🏦 How I Designed the BDA Synthetic Banking Simulator

---

# Welcome

Thank you for exploring the **BDA Synthetic Banking Simulator**.

Unlike many machine learning projects that begin with an existing dataset, this project started with a different question:

> **How can we create a realistic synthetic banking environment that supports ethical research, education, and machine learning without relying on confidential customer information?**

This notebook explains the design decisions behind the simulator, the assumptions that guided its development, and the research philosophy that shaped the project.

---

# Why I Built This Project

I created this project as part of my learning journey while completing my **Master of Science in Data Science**.

My long-term goal is to work as a **Data Scientist in the financial services industry**, with a particular interest in using data to improve customer care, accessibility, and service delivery.

Before beginning graduate school, I spent more than five years working as a bookkeeper, where I developed an appreciation for financial records, data quality, and how operational decisions affect businesses.

Today, I work in retail banking in a customer-facing role, where I regularly see how customers interact with different banking services and channels. Those experiences reinforced an idea that has become central to this project:

> Technology should make banking easier for customers, not more difficult.

Rather than viewing digital banking adoption as a way to reduce costs, I wanted to explore how data science could be used to identify customers who might benefit from additional education, onboarding, or personalized support.

This simulator reflects that philosophy.

# Why Synthetic Data?

Financial institutions manage highly sensitive customer information.

Because of privacy legislation, internal policies, and ethical responsibilities, these datasets are generally unavailable for public research.

Synthetic data offers an alternative.

Instead of copying real customer records, synthetic datasets simulate realistic patterns while protecting individual privacy.

This allows researchers to:

- prototype models
- teach data science
- test software
- validate workflows

without exposing personal information.

---

# Research Philosophy

One of the guiding principles behind this project is that:

> Realistic does not mean identical to reality.

The simulator is not intended to recreate any specific bank or customer.

Instead, it generates believable patterns that resemble how a retail banking environment may behave over time.

This distinction is important because it allows experimentation while avoiding the use of real customer information.

---

# Why BDA?

The fictional bank, **La Banque D'Avoine (BDA)**, was created specifically for this project.

The bank has:

- headquarters in Québec
- a growing presence in Vancouver
- retail and business customers
- a downtown branch serving a diverse population

Creating a fictional institution allows the simulator to explore operational questions without representing any real financial institution.

---

# Designing Customer Behaviour

One of the biggest challenges was avoiding completely random transactions.

Real customers behave differently.

Some primarily use:

- branches
- ATMs
- mobile banking
- online banking
- telephone banking

These preferences also evolve over time.

To represent this, each synthetic customer receives an internal behavioural profile that influences—not determines—their decisions.

The simulator uses probabilities rather than rigid rules, allowing for natural variation between customers.

---

# Simulating Time

Banking behaviour changes over time.

Rather than treating every day equally, the simulator includes:

- weekends
- public holidays
- seasonal effects
- weather events
- sporting events
- long-term behavioural change

These factors influence the likelihood of different banking activities.

---

# Modelling the COVID-19 Pandemic

The COVID-19 pandemic significantly altered banking behaviour around the world.

Rather than ignoring this period, the simulator models it as a behavioural transition.

Examples include:

- reduced branch traffic
- increased online banking
- increased telephone banking
- increased password resets
- gradual recovery after restrictions eased

The goal is not to reproduce exact historical statistics but to create a plausible environment for studying behavioural change.

Many discussions around banking technology focus on efficiency.

# Machine Learning Philosophy

The machine learning models developed using this simulator are designed to answer questions such as:

- Which factors influence digital adoption?
- How do external events affect banking behaviour?

Rather than optimizing only predictive accuracy, the project also emphasizes:

- interpretability
- fairness
- transparency
- reproducibility

These principles are especially important in financial services.

# Limitations

Like every simulation, this project has limitations.

The simulator cannot capture every aspect of human behaviour or banking operations.

Examples include:

- individual decision-making
- macroeconomic events
- regulatory changes
- fraud
- investment behaviour
- commercial lending

Instead, the simulator provides a simplified environment suitable for experimentation and education.
