# 🏦 Customizing Your Synthetic Bank

## Welcome back!

In the previous notebook, you generated your first synthetic banking dataset using the default settings for La Banque D'Avoine (BDA), a fictional bank made by me.

In this notebook, you'll learn how to customize your bank to create datasets that better fit your own research questions.

By the end of this notebook, you'll know how to:

* Change the number of customers
* Add or remove bank branches
* Modify transaction types
* Adjust account types
* Change the simulation period
* Create small or very large datasets

**No programming experience beyond basic Python is required.**

---

## Why Customize?

_No two banks are exactly alike._

A small community credit union behaves differently from a national bank. Likewise, a bank located in downtown Vancouver serves different customers than one in rural British Columbia.

Instead of hard-coding one type of bank, the BDA simulator is designed to be configurable so you can create datasets that better represent the institution you're interested in studying.

---

## Understanding config.py

Almost every setting used by the simulator is defined in config.py.

Think of this file as the "control panel" for your synthetic bank.

You'll typically modify:

* number of customers
* simulation dates
* branches
* transaction types
* customer behaviour assumptions

Changing these values allows the rest of the simulator to adapt automatically.

---

## Changing the Number of Customers

The simplest customization is changing how many customers your bank serves.

For example:

* 500 customers → classroom demonstrations
* 10,000 customers → portfolio projects
* 100,000+ customers → performance testing

After modifying the configuration, simply rerun the generator to create a new dataset.

---

## Adding New Branches

Suppose BDA (our whatever name you give your synthetic bank) opens a new branch in Burnaby(or wherever you choose).

You'll add the branch to your branch configuration with:

* a unique branch ID
* transit number
* branch name
* address
* city

Once added, new customers can be assigned to that branch, and transactions may occur there.

---

## Creating New Transaction Types

Banks _continually_ introduce new products and services.

For example, you may wish to simulate:

* Interac e-Transfers
* Foreign currency exchange
* Mortgage applications
* Investment contributions
* Wire transfers

When adding a new transaction type, ask yourself:

* Is money being moved?
* Which banking channels allow this transaction?
* Which customers are most likely to perform it?

Thinking through these questions helps create more realistic data.

---

## Modifying Account Types

Retail banks often offer many products.

Examples include:

* Chequing
* Savings
* TFSA
* RRSP
* FHSA
* Business Chequing
* Business Savings
* USD Savings

Adding new account types allows researchers to study how customers interact with different financial products over time.

---

## Changing the Simulation Period

The default simulator models behaviour between:

**January 1, 2019**

_and_

**December 31, 2024**

You can easily modify these dates to:

* simulate future banking trends
* study only the pandemic years
* create shorter datasets for testing

Longer simulations naturally produce larger transaction datasets.

---

## Experiment Ideas

Try creating several different banks:

**Community Credit Union**

* 2,000 customers
* 2 branches
* older population
* high branch usage

**Digital-First Bank**

* 50,000 customers
* no physical branches
* mobile-first behaviour

**Business Banking Simulation**

* high percentage of business accounts
* larger deposits
* more drafts and wire transfers

Each variation creates new research opportunities.
